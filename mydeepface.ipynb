**To Draw rectangle araound the face**

```python
image_path = './margot.jpg'
image = Image.open(image_path)


analysis = DeepFace.analyze(img_path=image_path,detector_backend='retinaface')

region = analysis[0]['region']
draw = ImageDraw.Draw(image)
draw.rectangle([(region['x'],region['y']),(region['x']+region['w'],region['y']+region['h'])],outline='green',width=5)

modified_image_path = './margot.jpg'

image.save(modified_image_path)


image = Image.open(modified_image_path)

resized = IImage(modified_image_path, width=300)
display(resized)
```
